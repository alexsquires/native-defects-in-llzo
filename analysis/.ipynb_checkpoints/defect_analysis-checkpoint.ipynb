### Chemical potential stability limits

This next cell plots the chemical potential stability regions from the output of [cplap](https://sourceforge.net/projects/cplap/). It reads in the `stability_grid.dat` file present in this directory as a csv, and plots a convex hull around the points in the grid to form a stabilty region.

```python
stability_region_grid = (read_csv('data/stability_grid.dat', skiprows=10, sep='\s+'))

rcParams.update( fp.stab_region_formatting )

fig, axs = plt.subplots(3,1, figsize =(6,14))
li_plot = axs[0]
la_plot = axs[1]
zr_plot = axs[2]


ax = li_plot
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((stability_region_grid['mu_O']),stability_region_grid['mu_Li'],color ='#ffffff', label = ' ', alpha=0)
ab.encircle((stability_region_grid['mu_O']),stability_region_grid['mu_Li'],fc="C0", ax=ax, alpha = 0.6, label='Li')
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ [eV]')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ [eV]')
ax.set_xlim(min(stability_region_grid['mu_O']),max(stability_region_grid['mu_O']))
ax.set_ylim(min(stability_region_grid['mu_Li']),max(stability_region_grid['mu_Li']))

ax = la_plot
ax.set_xticklabels([])
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((stability_region_grid['mu_O']),stability_region_grid['mu_La'],color ='#ffffff', label = ' ', alpha=0)
ab.encircle((stability_region_grid['mu_O']),stability_region_grid['mu_La'], ax=ax, fc="C1", alpha = 0.6, label='La')
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_ylabel('$\Delta\mu_\mathrm{La}$ [eV]')
ax.set_xlim(min(stability_region_grid['mu_O']),max(stability_region_grid['mu_O']))
ax.set_ylim(min(stability_region_grid['mu_La']),max(stability_region_grid['mu_La']))

ax = zr_plot
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((stability_region_grid['mu_O']),stability_region_grid['mu_Zr'],color ='#ffffff', label = ' ', alpha=0)
ab.encircle((stability_region_grid['mu_O']),stability_region_grid['mu_Zr'],fc="C2", ax=ax, alpha = 0.6, label='Zr')
ax.yaxis.tick_right()
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
ax.set_xticklabels([])
ax.yaxis.set_label_position('right') 
ax.set_ylabel('$\Delta\mu_\mathrm{Zr}$ [eV]')
ax.set_xlim(min(stability_region_grid['mu_O']),max(stability_region_grid['mu_O']))
ax.set_ylim(min(stability_region_grid['mu_Zr']),max(stability_region_grid['mu_Zr']))
plt.savefig('../figs/stability_region.pdf')
plt.show()
```

### Lithium off-stoichiometry over the chemical potential stability region

This cell takes a section of the stability region and calculates lithium off stoichiometry across them, this is achieved by taking the chemical potentials over each point in the grid as defined previously, and carrying out an sc-fermi calculation for that set of chemical potentials.

```python
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

stability_region_grid = (read_csv('data/stability_grid.dat', skiprows=10, sep='\s+'))

rcParams.update( fp.master_formatting )


cmap = sns.cubehelix_palette(8, start=.5, rot=-.75, light=1,as_cmap=True,reverse=False)

calcs = []
for o_limit,li_limit,la_limit,zr_limit in zip(stability_region_grid['mu_O'],
                                              stability_region_grid['mu_Li'],
                                              stability_region_grid['mu_La'],
                                              stability_region_grid['mu_Zr']):
    calc = ab.make_calcs(o_limit,li_limit,la_limit,zr_limit)
    calcs.append(calc)
results = []
for i in calcs:
    result = ab.run_some_fermi(i,1500)
    results.append(result)

# vacancy defects
li_vacancies_neutral = [i.li_vac_0 for i in results]
li_vacancies_n1 = [i.li_vac_1 for i in results]
o_vacancies_neutral = [i.o_vac_0 for i in results]
o_vacancies_p1 = [i.o_vac_1 for i in results]
o_vacancies_p2 = [i.o_vac_2 for i in results]
la_vacancies_p3 = [i.la_vac_3 for i in results]
zr_vacancies_p4 = [i.zr_vac_4 for i in results]
# interstitial defects
li_interstitals_neutral = [i.li_int_0 for i in results]
li_interstitals_p1 = [i.li_int_1 for i in results]
o_interstitals_neutral = [i.o_int_0 for i in results]
o_interstitals_n1 = [i.o_int_1 for i in results]
# cation subsitiution defects
la_zr_neutral = [i.la_zr_0 for i in results]
la_zr_n1 = [i.la_zr_1 for i in results]
zr_la_neutral = [i.zr_la_0 for i in results]
zr_la_p1 = [i.zr_la_1 for i in results]
zr_li_oct_neutral = [i.zr_li_0 for i in results]
zr_li_oct_p1 = [i.zr_li_1 for i in results]
zr_li_oct_p2 = [i.zr_li_2 for i in results]
zr_li_oct_p3 = [i.zr_li_3 for i in results]
zr_li_tet_neutral = [i.zr_li_tet_0 for i in results]
zr_li_tet_p1 = [i.zr_li_tet_1 for i in results]
zr_li_tet_p2 = [i.zr_li_tet_2 for i in results]
zr_li_tet_p3 = [i.zr_li_tet_3 for i in results] 

net = []               
for liv0,liv1,zrli0,zrli1,zrli2,zrli3,zrlitet0,zrlitet1,zrlitet2,zrlitet3,lint0,lint1 in zip(li_vacancies_neutral,
                                                                                             li_vacancies_n1,
                                                                                             zr_li_oct_neutral,
                                                                                             zr_li_oct_p1,
                                                                                             zr_li_oct_p2,
                                                                                             zr_li_oct_p3,
                                                                                             zr_li_tet_neutral,
                                                                                             zr_li_tet_p1,
                                                                                             zr_li_tet_p2,
                                                                                             zr_li_tet_p3,
                                                                                             li_interstitals_neutral,
                                                                                             li_interstitals_p1):
    value = (lint0+lint1) - (liv0+liv1+zrli0+zrli1+zrli2+zrli3+zrlitet0+zrlitet1+zrlitet2+zrlitet3)
    net.append(value)
    
fig, ax = plt.subplots()
axins = inset_axes(ax, 3,3 , loc=2, bbox_to_anchor=(.1, 0.4))

ax.tricontourf((stability_region_grid['mu_O']),(stability_region_grid['mu_Li']), 
                (np.nan_to_num(ab.off_stoich(np.array(net)))),9,cmap=cmap)#
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ [eV]', fontsize='18')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ [eV]', fontsize='18')

plot = axins.tricontourf((stability_region_grid['mu_O']),(stability_region_grid['mu_Li']), 
                (np.nan_to_num(ab.off_stoich(np.array(net)))),9,cmap=cmap)


ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 



axins.xaxis.tick_top()
axins.yaxis.tick_right()
axins.xaxis.set_label_position('top')
axins.yaxis.set_label_position('right') 
axins.set_xlim(-1.7, -1.2)
axins.set_ylim(-1.9, -1.7)

cbax = fig.add_axes([1.07, 0.14, 0.05, 0.7])
cbar = fig.colorbar(plot, cax=cbax)
cbar.ax.tick_params(labelsize=13) 
cbar.ax.set_ylabel( 'Off-stoichiometry [Li / Unit Cell]', fontsize='18' )
cbar.outline.set_visible(False)


mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5", zorder=5)

plt.savefig(fname='../figs/off_stoich.pdf', bbox_inches='tight', dpi = 600)
plt.show()
        
```

### Defect concentrations as a function of temperature

This cell determines defect concentrations as a function of temperature. This is done via the procedure as discussed in the paper. 

```python
import numpy as np
from matplotlib import rc, rcParams
from collections import OrderedDict
rcParams.update( fp.master_formatting )


fig, ax = plt.subplots()
t = np.arange(700,1510,10)


v = []
for i in t:
    g = ab.give_me_concs_sometimes(1e1,i)
    v.append(g)

vo = [i[0]['vo2'] + i[0]['vo1'] + i[0]['vo0'] for i in v]
io = [i[0]['io1'] + i[0]['io0'] for i in v]
vli = [i[0]['vli1'] + i[0]['vli0'] for i in v]
ili = [i[0]['ili1'] + i[0]['ili0'] for i in v]
zrlioct = [i[0]['zrlioct0'] + i[0]['zrlioct1'] + i[0]['zrlioct2'] +i[0]['zrlioct3'] for i in v]
zrlitet = [i[0]['zrlitet0'] + i[0]['zrlitet1'] + i[0]['zrlitet2'] +i[0]['zrlitet3'] for i in v]
vla = [i[0]['vla3'] for i in v]
vzr = [i[0]['vzr4'] for i in v]
lazr = [i[0]['lazr0'] + i[0]['lazr1'] for i in v] 
zrla = [i[0]['zrla0'] + i[0]['zrla1'] for i in v]
d = [i[0]['mu_O'] for i in v]

ax.set_xlabel('T [K]')
ax.set_xlim(700,1500)
mn, mx = ax.get_ylim()
ax.set_ylabel('[defect] / $\log_{10}$ cm$^{-3}$' )
ax2 = ax.twinx()

ax.plot(t,np.log10(vo), color='C0', ls = '-', label = '$V_\mathrm{O}$', linewidth =2)
ax.plot(t,np.log10(vla), color='C1',  ls = '-', label = '$V_\mathrm{La}$')
ax.plot(t,np.log10(vzr), color='C2',  ls = '-', label = '$V_\mathrm{Zr}$', linewidth =2)
ax.plot(t,np.log10(lazr), color='C3',  ls = '-', label = '$\mathrm{La}_\mathrm{Zr}$', linewidth =2)
ax.plot(t,np.log10(zrla), color='C4',  ls = '-', label = '$\mathrm{Zr}_\mathrm{La}$', linewidth =2)
ax.plot(t,np.log10(zrlioct), color='C5',  ls = '-', label = '$\mathrm{Zr}^\mathrm{oct}_\mathrm{Li}$', linewidth =2)
ax.plot(t,np.log10(zrlitet), color='C6',  ls = '-', label = '$\mathrm{Zr}^\mathrm{tet}_\mathrm{Li}$', linewidth =2)
ax.plot(t,np.log10(io), color='C7',  ls = '-', label = '$\mathrm{O}_\mathrm{i}$')



ax2.set_ylabel('[defect] / $\log_{10}$ Unit Cell$^{-1}$')

ax.set_ylim(0,18)
ax.set_yticks(np.linspace(ax.get_ybound()[0], ax.get_ybound()[1], 5))
ax2.set_yticks(np.linspace(ax2.get_ybound()[0], ax2.get_ybound()[1], 5))
y_ticks = ((ab.off_stoich(10**ax.get_yticks())))
ax2.set_yticklabels(np.around(np.log10(y_ticks),2))


ax2.legend(bbox_to_anchor=(1.25, 0.9))
plt.savefig('../figs/all_defect_concs.pdf')
plt.show()
```
