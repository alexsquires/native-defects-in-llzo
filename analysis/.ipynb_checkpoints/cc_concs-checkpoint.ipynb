## Stability regions with respect to Li, Zr and La as a function of O.

In this next cell, staiblity regions are plotted with respect to oxygen, and each metal. This is acheived by plotting the stability information as plotted in in the `x_grid.dat` files as a scatter plot, and then surrounded with a convex hull.

```python
fig, axs = plt.subplots(3,1, figsize =(6,14))

li_plot = axs[0]
la_plot = axs[1]
zr_plot = axs[2]

ax = li_plot
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((grid['mu_O']),grid['mu_Li'],color ='#ffffff', label = ' ', alpha=0)
encircle((grid['mu_O']),grid['mu_Li'],fc="C0", ax=ax, alpha = 0.6, label='Li')
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ / eV')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_Li']),max(grid['mu_Li']))

ax = la_plot
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((grid['mu_O']),grid['mu_La'],color ='#ffffff', label = ' ', alpha=0)
encircle((grid['mu_O']),grid['mu_La'], ax=ax, fc="C1", alpha = 0.6, label='La')
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_ylabel('$\Delta\mu_\mathrm{La}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_La']),max(grid['mu_La']))

ax = zr_plot
ax.scatter((grid['mu_O']),grid['mu_Zr'],color ='#ffffff', label = ' ', alpha=0)
encircle((grid['mu_O']),grid['mu_Zr'],fc="C2", ax=ax, alpha = 0.6, label='Zr')
ax.yaxis.tick_right()
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_ylabel('$\Delta\mu_\mathrm{Zr}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_Zr']),max(grid['mu_Zr']))
plt.savefig('../figs/stability_region.pdf')
plt.show()
```

## Defect Chemistry as a function of oxygen chemical potential

For three different reigimes, an O rich, O poor and a moderate O chemical potential reigime as set in examples, we calculate self consistent defect formation energies and transition level diagrams for each of these sets of chemical potentials. Each is plotted with a labelled stability plot illustaing at what chemical potentials these values are calculated at. 

This is acheived by:  
- selecting three different sets of chemical potentials
- reading in the DFT data contained in the `x.yaml` files  
- running `sc_fermi` calculations and scraping out the concentrations using the scripts in `sc_pyper_functions.py`  
- reading in the `transition_levels.dat` file produced by `sc-fermi` and plotting this information 


```python
# The values in examples can be varied for other integers between 0 and 82 to investigate different sets of chemical potentials

examples =  (75,12,42) # define three points on the stabilty region grid that represent a low O chemical potential
                       # moderate O chemical potential, and high O chemical potential.  

fig, axs = plt.subplots(3,3, figsize =(18,18)) # define full figure
    
# plot stability region for first case    

ax = axs[0,0]
ax.set_title('Oxygen Poor/Metal Rich', pad=80)
ax.scatter((grid['mu_O']),grid['mu_Li'],color ='#ffffff', label = ' ', alpha=0)
ax.scatter(grid['mu_O'][examples[0]],grid['mu_Li'][examples[0]], color ='white',zorder=1)
encircle((grid['mu_O']),grid['mu_Li'],fc="black", ax=ax, alpha = 0.6, label='Li',zorder = -1)
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ / eV')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_Li']),max(grid['mu_Li']))

ax = axs[1,0]

results = run_some_fermi(compile_defects(grid.iloc[examples[0]].to_dict()), # read in chemical potentials from stablity region
                         nelect=544, # number of electrons in calculation
                         e_gap=5.9034, # bandgap magnitude
                         T=1500,       # temperature
                         spin_polarised=True) # is calculation spin polarised?

## Sum individual charge states of each defect under one variable

v_li = (results['v_Li_-1.0'] + results['v_Li_0.0'])
i_li = (results['Li_i_1.0'] + results['Li_i_0.0'])
v_la = (results['v_La_-3.0'])
v_zr = (results['v_Zr_-4.0'])
v_o = (results['v_O_0.0']+results['v_O_1.0']+results['v_O_2.0'])
o_i = (results['O_i_0.0']+results['O_i_-1.0'])
zr_li = (results['Zr_Li_0.0'] + results['Zr_Li_1.0']+results['Zr_Li_2.0']+results['Zr_Li_3.0'])
zr_li_t = ( results['Zr_Li_tet_0.0'] + results['Zr_Li_tet_1.0']+results['Zr_Li_tet_2.0']+results['Zr_Li_tet_3.0'])
zr_i = (results['Zr_i_0.0'] + results['Zr_i_1.0'] + results['Zr_i_2.0'] + results['Zr_i_3.0'] + results['Zr_i_4.0'])
la_zr = (results['La_Zr_0.0'] + results['La_Zr_-1.0'])
zr_la = (results['Zr_La_0.0'] + results['Zr_La_1.0'])
net_li = (v_li + zr_li - i_li)
if net_li < 0:
    net_li = net_li*-1
    c_net_li = c_li_i
else:
    c_net_li = c_v_li

# Plot defect concentrations on a bar chart
chart_labels= ('$V_\mathrm{Li}$', '$V_\mathrm{La}$', '$V_\mathrm{Zr}$', '$V_\mathrm{O}$','$\mathrm{O}_\mathrm{i}$',
                   '$\mathrm{Zr}^\mathrm{oct}_\mathrm{Li}$','$\mathrm{Zr}^\mathrm{tet}_\mathrm{Li}$','$\mathrm{Zr}_\mathrm{i}$',
                   '$\mathrm{La}_\mathrm{Zr}$','$\mathrm{Zr}_\mathrm{La}$','$\mathrm{e}^-$','$\mathrm{h}^+$')
defect_concs = np.log10([net_li,v_la,v_zr,v_o,o_i,zr_li,zr_li_t,zr_i,la_zr,zr_la,results['c_e'],results['c_h']])
index = np.arange(len(defect_concs))
ax.bar(index,defect_concs,color=(c_v_li,c_v_la,c_v_zr,c_v_o,c_o_i,c_zr_li_o,c_zr_li_t,c_zr_i,c_la_zr,c_zr_la,'#262728','gold'),width=0.75)
ax.set_xticks(index)
ax.set_xticklabels(chart_labels, fontsize=12)
ax.set_ylabel('Defect concentration / log$_{10}$ cm$^{-3}$')
ax.set_ylim([0,None])

# Read transition levels from sc-fermi output and plot

ax = axs[2,0]
transition_levels = read_full_transition_levels()
ax.plot(transition_levels[0]['#'],transition_levels[0]['v_O'], '-o',label = '$V_\mathrm{O}$', color = c_v_o, markersize = 9)                
ax.plot(transition_levels[1]['#'],transition_levels[1]['v_Li'], '-o',label = '$V_\mathrm{Li}$', color= c_v_li, markersize = 9)
ax.plot(transition_levels[2]['#'],transition_levels[2]['v_La'], '-o',label = '$V_\mathrm{La}$', color=c_v_la, markersize = 9)
ax.plot(transition_levels[3]['#'],transition_levels[3]['v_Zr'], '-o',label = '$V_\mathrm{Zr}$', color=c_v_zr, markersize = 9)
ax.plot(transition_levels[4]['#'],transition_levels[4]['Li_i'], '-o',label = '$\mathrm{Li}_\mathrm{i}$', color=c_li_i, markersize = 9)
ax.plot(transition_levels[5]['#'],transition_levels[5]['Zr_i'], '-o' ,label = '$\mathrm{Zr_{i}}$', color=c_zr_i, markersize = 9)
ax.plot(transition_levels[6]['#'],transition_levels[6]['O_i'], '-o',label = '$\mathrm{O_{i}}$', color=c_o_i, markersize = 9)
ax.plot(transition_levels[7]['#'],transition_levels[7]['La_Zr'], '-o',label = '$\mathrm{La_{Zr}}$', color=c_la_zr, markersize = 9)
ax.plot(transition_levels[8]['#'],transition_levels[8]['Zr_La'], '-o',label = '$\mathrm{Zr_{La}}$', color=c_zr_la, markersize = 9)
ax.plot(transition_levels[9]['#'],transition_levels[9]['Zr_Li'], '-o',label = '$\mathrm{Zr^{oct}_{Li}}$', color=c_zr_li_o, markersize = 9)
ax.plot(transition_levels[10]['#'],transition_levels[10]['Zr_Li_tet'], '-o',label = '$\mathrm{Zr^{tet}_{Li}}$', color=c_zr_li_t, markersize = 9)
ax.set_xlim(0,6)
ax.set_ylim(0,8)
ax.set_ylabel('Formation Energy / eV' )
ax.set_xlabel('Fermi Energy / eV' )
plt.tight_layout()

## Repeat process for other two sets of chemical potentials in examples:

ax = axs[0,2]
ax.set_title('Oxygen Rich/Metal Poor', pad =80)
ax.scatter((grid['mu_O']),grid['mu_Li'],color ='#ffffff', label = ' ', alpha=0)
ax.scatter(grid['mu_O'][examples[1]],grid['mu_Li'][examples[1]], color ='white',zorder=1)
encircle((grid['mu_O']),grid['mu_Li'],fc="black", ax=ax, alpha = 0.6, label='Li',zorder = -1)
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ / eV')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_Li']),max(grid['mu_Li']))


ax = axs[1,2]
results = run_some_fermi(compile_defects(grid.iloc[examples[1]].to_dict()), # read in chemical potentials from stablity region
                         nelect=544,   # number of electrons in calculation
                         e_gap=5.9034, # bandgap magnitude
                         T=1500,       # temperature
                         spin_polarised=True) # is calculation spin polarised?

v_li = (results['v_Li_-1.0'] + results['v_Li_0.0'])
i_li = (results['Li_i_1.0'] + results['Li_i_0.0'])
v_la = (results['v_La_-3.0'])
v_zr = (results['v_Zr_-4.0'])
v_o = (results['v_O_0.0']+results['v_O_1.0']+results['v_O_2.0'])
o_i = (results['O_i_0.0']+results['O_i_-1.0'])
zr_li = (results['Zr_Li_0.0'] + results['Zr_Li_1.0']+results['Zr_Li_2.0']+results['Zr_Li_3.0'])
zr_li_t = ( results['Zr_Li_tet_0.0'] + results['Zr_Li_tet_1.0']+results['Zr_Li_tet_2.0']+results['Zr_Li_tet_3.0'])
zr_i = (results['Zr_i_0.0'] + results['Zr_i_1.0'] + results['Zr_i_2.0'] + results['Zr_i_3.0'] + results['Zr_i_4.0'])
la_zr = (results['La_Zr_0.0'] + results['La_Zr_-1.0'])
zr_la = (results['Zr_La_0.0'] + results['Zr_La_1.0'])
net_li = (v_li + zr_li - i_li)
if net_li < 0:
    net_li = net_li*-1
    c_net_li = c_li_i
else:
    c_net_li = c_v_li


defect_concs = np.log10([net_li,v_la,v_zr,v_o,o_i,zr_li,zr_li_t,zr_i,la_zr,zr_la,results['c_e'],results['c_h']])
ind = np.arange(len(defect_concs))
ax.bar(ind, defect_concs ,color=(c_v_li,c_v_la,c_v_zr,c_v_o,c_o_i,c_zr_li_o,c_zr_li_t,c_zr_i,c_la_zr,c_zr_la,'#262728','gold'), width=0.75)
ax.set_xticks(ind)
ax.set_xticklabels(chart_labels, fontsize=12)
ax.set_ylabel('Defect concentration / log$_{10}$ cm$^{-3}$')

ax = axs[2,2]
transition_levels = read_full_transition_levels()
ax.plot(transition_levels[0]['#'],transition_levels[0]['v_O'], '-o',label = '$V_\mathrm{O}$', color = c_v_o, markersize = 9)                
ax.plot(transition_levels[1]['#'],transition_levels[1]['v_Li'], '-o',label = '$V_\mathrm{Li}$', color= c_v_li, markersize = 9)
ax.plot(transition_levels[2]['#'],transition_levels[2]['v_La'], '-o',label = '$V_\mathrm{La}$', color=c_v_la, markersize = 9)
ax.plot(transition_levels[3]['#'],transition_levels[3]['v_Zr'], '-o',label = '$V_\mathrm{Zr}$', color=c_v_zr, markersize = 9)
ax.plot(transition_levels[4]['#'],transition_levels[4]['Li_i'], '-o',label = '$\mathrm{Li}_\mathrm{i}$', color=c_li_i, markersize = 9)
ax.plot(transition_levels[5]['#'],transition_levels[5]['Zr_i'], '-o' ,label = '$\mathrm{Zr_{i}}$', color=c_zr_i, markersize = 9)
ax.plot(transition_levels[6]['#'],transition_levels[6]['O_i'], '-o',label = '$\mathrm{O_{i}}$', color=c_o_i, markersize = 9)
ax.plot(transition_levels[7]['#'],transition_levels[7]['La_Zr'], '-o',label = '$\mathrm{La_{Zr}}$', color=c_la_zr, markersize = 9)
ax.plot(transition_levels[8]['#'],transition_levels[8]['Zr_La'], '-o',label = '$\mathrm{Zr_{La}}$', color=c_zr_la, markersize = 9)
ax.plot(transition_levels[9]['#'],transition_levels[9]['Zr_Li'], '-o',label = '$\mathrm{Zr^{oct}_{Li}}$', color=c_zr_li_o, markersize = 9)
ax.plot(transition_levels[10]['#'],transition_levels[10]['Zr_Li_tet'], '-o',label = '$\mathrm{Zr^{tet}_{Li}}$', color=c_zr_li_t, markersize = 9)
ax.set_xlim(0,6)
ax.set_ylim(0,8)
ax.set_ylabel('Formation Energy / eV' )
ax.set_xlabel('Fermi Energy / eV' )
plt.tight_layout()


ax = axs[0,1]
ax.set_title('Mid Point', pad = 80)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.scatter((grid['mu_O']),grid['mu_Li'],color ='#ffffff', label = ' ', alpha=0)
ax.scatter(grid['mu_O'][examples[2]],grid['mu_Li'][examples[2]], color ='white',zorder=1)
encircle((grid['mu_O']),grid['mu_Li'],fc="black", ax=ax, alpha = 0.6, label='Li',zorder = -1)
ax.xaxis.tick_top()
ax.yaxis.tick_right()
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right') 
ax.set_xlabel('$\Delta\mu_\mathrm{O}$ / eV')
ax.set_ylabel('$\Delta\mu_\mathrm{Li}$ / eV')
ax.set_xlim(min(grid['mu_O']),max(grid['mu_O']))
ax.set_ylim(min(grid['mu_Li']),max(grid['mu_Li']))


ax = axs[1,1]
results = run_some_fermi(compile_defects(grid.iloc[examples[2]].to_dict()), # read in chemical potentials from stablity region
                         nelect=544,   # number of electrons in calculation
                         e_gap=5.9034, # bandgap magnitude
                         T=1500,       # temperature
                         spin_polarised=True) # is calculation spin polarised?

v_li = (results['v_Li_-1.0'] + results['v_Li_0.0'])
i_li = (results['Li_i_1.0'] + results['Li_i_0.0'])
v_la = (results['v_La_-3.0'])
v_zr = (results['v_Zr_-4.0'])
v_o = (results['v_O_0.0']+results['v_O_1.0']+results['v_O_2.0'])
o_i = (results['O_i_0.0']+results['O_i_-1.0'])
zr_li = (results['Zr_Li_0.0'] + results['Zr_Li_1.0']+results['Zr_Li_2.0']+results['Zr_Li_3.0'])
zr_li_t = ( results['Zr_Li_tet_0.0'] + results['Zr_Li_tet_1.0']+results['Zr_Li_tet_2.0']+results['Zr_Li_tet_3.0'])
zr_i = (results['Zr_i_0.0'] + results['Zr_i_1.0'] + results['Zr_i_2.0'] + results['Zr_i_3.0'] + results['Zr_i_4.0'])
la_zr = (results['La_Zr_0.0'] + results['La_Zr_-1.0'])
zr_la = (results['Zr_La_0.0'] + results['Zr_La_1.0'])
net_li = (v_li + zr_li - i_li)
if net_li < 0:
    net_li = net_li*-1
    c_net_li = c_li_i
else:
    c_net_li = c_v_li

defect_concs = np.log10([net_li,v_la,v_zr,v_o,o_i,zr_li,zr_li_t,zr_i,la_zr,zr_la,results['c_e'],results['c_h']])
ax.bar(index, defect_concs, color=(c_v_li,c_v_la,c_v_zr,c_v_o,c_o_i,c_zr_li_o,c_zr_li_t,c_zr_i,c_la_zr,c_zr_la,'#262728','gold'), width=0.75)
ax.set_xticks(index)
ax.set_xticklabels(chart_labels,fontsize=12)
ax.set_ylabel('Defect concentration / log$_{10}$ cm$^{-3}$')


ax = axs[2,1]
transition_levels = read_full_transition_levels()
ax.plot(transition_levels[0]['#'],transition_levels[0]['v_O'], '-o',label = '$V_\mathrm{O}$', color = c_v_o, markersize = 9)                
ax.plot(transition_levels[1]['#'],transition_levels[1]['v_Li'], '-o',label = '$V_\mathrm{Li}$', color= c_v_li, markersize = 9)
ax.plot(transition_levels[2]['#'],transition_levels[2]['v_La'], '-o',label = '$V_\mathrm{La}$', color=c_v_la, markersize = 9)
ax.plot(transition_levels[3]['#'],transition_levels[3]['v_Zr'], '-o',label = '$V_\mathrm{Zr}$', color=c_v_zr, markersize = 9)
ax.plot(transition_levels[4]['#'],transition_levels[4]['Li_i'], '-o',label = '$\mathrm{Li}_\mathrm{i}$', color=c_li_i, markersize = 9)
ax.plot(transition_levels[5]['#'],transition_levels[5]['Zr_i'], '-o' ,label = '$\mathrm{Zr_{i}}$', color=c_zr_i, markersize = 9)
ax.plot(transition_levels[6]['#'],transition_levels[6]['O_i'], '-o',label = '$\mathrm{O_{i}}$', color=c_o_i, markersize = 9)
ax.plot(transition_levels[7]['#'],transition_levels[7]['La_Zr'], '-o',label = '$\mathrm{La_{Zr}}$', color=c_la_zr, markersize = 9)
ax.plot(transition_levels[8]['#'],transition_levels[8]['Zr_La'], '-o',label = '$\mathrm{Zr_{La}}$', color=c_zr_la, markersize = 9)
ax.plot(transition_levels[9]['#'],transition_levels[9]['Zr_Li'], '-o',label = '$\mathrm{Zr^{oct}_{Li}}$', color=c_zr_li_o, markersize = 9)
ax.plot(transition_levels[10]['#'],transition_levels[10]['Zr_Li_tet'], '-o',label = '$\mathrm{Zr^{tet}_{Li}}$', color=c_zr_li_t, markersize = 9)
ax.set_xlim(0,6)
ax.set_ylim(0,8)
ax.set_ylabel('Formation Energy / eV' )
ax.set_xlabel('Fermi Energy / eV' )
plt.tight_layout()

#final plot and save figure for paper

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('../figs/chems_plus.pdf')
plt.show()
```

## Defects as a function of lithium chemical potential

After exploring how defect chemistry can change as a function of oxygen chemical potential, we can take a fixed oxygen chemical potential and see how defect chemistry changes as a function of lithium chemical potential. This is done by taking the `ex_grid.dat` and taking some oxygen chemical potential value, finding the relevent lithium chemical potentials at the value and for each of them, calculating a range of defect concentrations, and seeing how they vary as lithium chemical potential is varied. We can exploit the `frozen-sc-fermi` program to investigate how the system might respond to dopants at different chemical potentials. To do this, we 'freeze in' the concentration of a dopant and calculate different defect concentrations as we change the concentration of the dopant.

```python
class chg_car_SCFermi:
    """
    SCFermi class containing all information needed to carry out an SC-Fermi calculationm including defects, number of eletrons, band gap magnitude, temperature, and whether the DFT calculation was spin polarised 
    currently only allows consideration of one 'dopant'
    """

    def __init__( self, defects, nelect, e_gap, temperature, dopants=None, fixed_defects=None, spin_polarised=False):
        self.defects = defects
        self.fixed_defects = fixed_defects
        self.nelect = nelect
        self.e_gap = e_gap
        self.temperature = temperature
        self.spin_polarised = spin_polarised
        self.dopants = dopants

    @property
    def n_defects( self ):
        return len( self.defects )
    
    @property
    def n_fixed_defects( self ):
        return len( self.fixed_defects )

    def output( self ):
            with open('input-fermi-frozen.dat', 'w') as f:
                if self.spin_polarised:
                    f.write( '2' + '\n')
                else:
                    f.write( '1' + '\n' )
                f.write( str(self.nelect) + '\n' )
                f.write( str(self.e_gap) + '\n')
                f.write( str(self.temperature) + '\n')
                f.write( str(self.n_defects) + '\n' )
                for d in self.defects:
                    f.write( '{} {} {}'.format( d.label, d.n_charge_states, d.n_sites ) + '\n')
                    for c in d.charge_states:
                        f.write( '{} {} {}'.format( c.charge, c.formation_energy, c.degeneracy ) + '\n')
                
                if self.fixed_defects is not None:
                    f.write( str(self.n_fixed_defects) + '\n' )
                    for k,v in self.fixed_defects.items():
                        f.write( '{} {}'.format( k, v ) + '\n')
                else:
                    f.write('0' + '\n')
                if self.dopants is not None:
                    f.write( '1' + '\n')
                    for k,v in self.dopants.items():
                        f.write( 'Dopant ' + '{} {}'.format( k, v ) + '\n')
                else:
                    f.write('0')
            f.close()
            
    
def run_some_chg_car_fermi(defects, nelect, e_gap, temperature, dopants=None, fixed_defects=None, spin_polarised=False):
    """
    a function to run a frozen SCFermi calculation, and retun the concentrations of all the relevent defects. 
    args:
        defects (list): a list of defects to be considered as part of this calculation
        T (float): the temperature (in Kelvin) the concentrations are to be calculated at
        e_gap (float): magnitude of bandgap
        spin_polarised (bool): whether the DFT data is spin polarised
        dopant_chg (int): charge on the frozen defect
        dopant_conc (int): concentration of the frozen defect in n/cm-3
    returns:
        flat (dict): a dictionary specifiying defect concentrations.
        
        Note: function currently assumes executable frozen sc-fermi in directory.
    """
    out = []
    scf = chg_car_SCFermi(defects, nelect, e_gap, temperature, dopants, fixed_defects, spin_polarised=False)
    scf.output()
    with open("out.txt", 'w') as f:
        sp = subprocess.run(["../../frozen-sc-fermi"],stdout=f)
        text_file = open(("out.txt") , "r")
        lines =  text_file.readlines()
        for i in defects:
                for line in lines:
                    if re.search(str(i.label)+' '+r'.*?Charge',line) is not None:
                        for h in range(i.n_charge_states):
                            joop = lines[lines.index(line)+(h+1)]
                            coop = joop.split()
                            x = coop[1]
                            y = float(coop[-2])
                            a_dict = {(i.label)+'_'+str(x) : y}
                            out.append(a_dict)
                            flat = {k: v for d in out for k, v in d.items()}
        for line in lines:
            if re.search('\(electrons\)', line) is not None:
                flat.update({'c_e' : float(line.split()[3])})
            if re.search('\(holes\)', line) is not None:
                flat.update({'c_h' : float(line.split()[3])})
            if re.search('\(eV\)', line) is not None:
                flat.update({'Fermi_level' : float(line.split()[4])})
            if re.search('Dopant\ \ \ \ \ \ \ \ \ :', line) is not None:
                flat.update({'Dopant' : float(line.split()[2])})
        return flat

itscrazybutitmightjustwork = []        
for i in range(len(grid)): 
    a = make_defect([defects['v_O'],defects['v_O_+'],defects['v_O_++']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs, sites=3 )
    b = make_defect([defects['Li_i'],defects['Li_i_+']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs )
    c = make_defect([defects['v_Li'],defects['v_Li_-']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs, sites=3 )
    d = make_defect([defects['v_La_---']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs, sites=2 )
    e = make_defect([defects['v_Zr_----']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs )
    f = make_defect([defects['Zr_La'],defects['Zr_La_+']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict() , corr=iccs, sites=2)
    g = make_defect([defects['La_Zr_-'],defects['La_Zr']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs)
    h = make_defect([defects['Zr_Li'],defects['Zr_Li_+'],defects['Zr_Li_++'],defects['Zr_Li_+++']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs,sites=2 )
    j = make_defect([defects['O_i'],defects['O_i_-']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict() , corr=iccs)
    p = make_defect([defects['Zr_Li_tet'],defects['Zr_Li_tet_+'],defects['Zr_Li_tet_++'],defects['Zr_Li_tet_+++']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs )
    q = make_defect([defects['Zr_i'],defects['Zr_i_+'],defects['Zr_i_++'],defects['Zr_i_+++'],defects['Zr_i_++++']], elements, interest['LLZO'], delta_mu=grid.iloc[i].to_dict(), corr=iccs )
    results = run_some_fermi([a,b,c,d,e,f,g,h,j,p,q],nelect=544, e_gap=5.9034, T=1500, spin_polarised=True)
    for_frozen = {k : v for k,v in results.items() if k is not 'Fermi_level' and k is not 'c_e' and k is not 'c_h'}
    formatt = [(str(k.split('_')[:-1]), v) for k,v in for_frozen.items()]
    z = defaultdict(list) 
    for k, v in formatt:
        z[k].append(v)
    gg = {k:sum(v) for k,v in z.items()}
    regex = re.compile('[^a-zA-Z,]')
    x = {regex.sub('', k):v for k,v in gg.items()}
    regex = re.compile('[^a-zA-Z]')
    y = {regex.sub('_', k):v for k,v in x.items()}
    please = run_some_chg_car_fermi([a,b,c,d,e,f,g,h,j,p,q],nelect=544, e_gap=5.9034, temperature=400, spin_polarised=True, fixed_defects=y, dopants={int(2) : float(per_form_un(wow))})
    itscrazybutitmightjustwork.append(please)      
```
